```python
INICIO
  f=2, rut, digito, m, sumatoria=0, mod11, candidato :entero
  DV :texto

  ESCRIBIR "Ingrese su cadena de rut"
  LEER rut

  MIENTRAS rut>0
    digito=rut%10
    rut=rut//10

    m=digito*f
    f=f+1

    SI F>7 ENTONCES
      f=2

    sumatoria=sumatoria+m

  mod11=sumatoria%11
  candidato=11-mod11

  SI candidato<10 ENTONCES
    DV=candidato

  SI candidato=10 ENTONCES
    DV="K"

  SI candidato=11 ENTONCES
    DV=0

  ESCRIBIR "Su DV es ",DV  
    
FIN
```

```python
# decla.var.
#  f=2, rut, digito, m, sumatoria=0, mod11, candidato :entero
#  DV :texto

f=2
sumatoria=0

rut=int(input("Ingrese su cadena de rut "))

while rut>0:
  digito=rut%10
  rut=rut//10
  m=digito*f
  f=f+1

  if f>7:
    f=2

  sumatoria=sumatoria+m

mod11=sumatoria%11
candidato=11-mod11

if candidato<10:
  DV=candidato
elif candidato==10:
  DV="K"
else:
  DV=0

print("Su DV es",DV)  
    
```
